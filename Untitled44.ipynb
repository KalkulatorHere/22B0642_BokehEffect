```python
import pandas,numpy
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import torchvision
import torch
import torchvision.datasets as datasets
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
```

```python
train_data = datasets.MNIST(root='./data', train=True, download=True,transform=transform)
print(len(train_data))
```

```python
train_loader = torch.utils.data.DataLoader(train_data, batch_size=64, shuffle=True)

images, labels = next(iter(train_loader))

plt.figure(figsize=(15, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(images[i].numpy().squeeze(), cmap='gray_r')
    plt.title(f"Label: {labels[i].item()}")
    plt.axis('off')

plt.show()
```

```python
data = torch.cat([image.view(1, -1) for image, _ in train_data], dim=0)
labels = torch.tensor([label for _, label in train_data])

data = data / 255.0

data_np = data.numpy()
labels_np = labels.numpy()
```

```python
import pandas as pd
label_counts = pd.Series(labels_np).value_counts().sort_index()
print("Label Distribution:")
print(label_counts)
plt.bar(label_counts.index, label_counts.values, color='black')
plt.title('Label Distribution')
plt.xlabel('Digit')
plt.ylabel('Count')
plt.xticks(label_counts.index)
plt.show()
```

```python
fig, axs = plt.subplots(2, 5, figsize=(10, 4))
for i in range(10):
    digit_images = data_np[labels_np == i][:5]
    for j in range(5):
        axs[j // 5, j % 5].imshow(digit_images[j].reshape(28, 28), cmap='gray')
        axs[j // 5, j % 5].axis('off')
        axs[j // 5, j % 5].set_title(f"Digit {i}")

plt.show()
```
